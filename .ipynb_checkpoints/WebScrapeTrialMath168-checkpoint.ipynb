```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

cbb_teams = pd.read_csv("NCAAD1BBall_Programs_teams.csv", header = None)
cbb_list = cbb_teams[0]

cbb_graph = nx.Graph()
for i in range(1, len(cbb_list)):
    cbb_graph.add_node(cbb_list[i])

print(cbb_graph)
```

```python
nx.draw(cbb_graph, with_labels = True)
plt.show
```

```python
import numpy as np

cbb_adjacency = pd.read_csv("NCAAD1BBall_Programs_Adjacency.csv")
cbb_adjacency.set_index("Unnamed: 0", inplace = True)
cbb_adjacency.index.name = "Team Name"
team_names = cbb_adjacency.index
diagonal = np.diag_indices(len(cbb_adjacency))
cbb_adjacency.iloc[diagonal] = np.zeros(len(cbb_adjacency))
cbb_adjacency
```
